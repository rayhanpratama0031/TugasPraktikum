```python
# Import module
import pandas as pd
import seaborn as sns
from google.colab import files

# Upload file dari lokal
uploaded = files.upload()
```

```python
# Load the CSV file
data1 = pd.read_csv("nba.csv")

# Show first 5 rows
data1.head()
```

```python
sns.barplot(x ="Age", y ="Weight", data = data1)
```

```python
sns.pointplot(x = "Age", y = "Weight", data = data1)
```

```python
 sns.boxplot(x=data1['Age'], y=data1['Weight'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot: Weight terhadap Age, dikelompokkan berdasarkan Position
sns.scatterplot(x=data1['Age'], y=data1['Weight'], hue=data1["Position"])

plt.title("Scatter Plot: Age vs Weight by Position")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.legend(title='Position')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Line plot: Weight terhadap Age, dikelompokkan berdasarkan Position
sns.lineplot(x=data1['Age'], y=data1['Weight'], hue=data1["Position"])

plt.title("Line Plot: Age vs Weight by Position")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.legend(title='Position')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Line plot antara Age dan Salary
sns.lineplot(x=data1["Age"], y=data1["Salary"])

plt.title("Line Plot: Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
sns.scatterplot(x=data1['Age'], y=data1['Salary'])

plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat lineplot dari kolom "Salary"
sns.lineplot(x=data1.index, y=data1["Salary"])

# Menampilkan plot
plt.title("Line Plot of Salary")
plt.xlabel("Index")
plt.ylabel("Salary")
plt.show()
```
