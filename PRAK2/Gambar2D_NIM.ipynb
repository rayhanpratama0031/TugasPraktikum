```python
"""
CIRCLES
"""

import numpy as np
import matplotlib.pyplot as plt

plt.axis([-75, 75, 50, -50])

plt.axis('on')
plt.grid(True)

plt.arrow(0, 0, 20, 0, head_length=4, head_width=3, color='k')
plt.arrow(0, 0, 0, 20, head_length=4, head_width=3, color='k')

plt.text(16, -3, 'x')
plt.text(-5, 17, 'y')

# _________________________________green circle
xc = 0
yc = 0
r = 40

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100
xlast = xc + r * np.cos(p1)
ylast = yc + r * np.sin(p1)
for p in np.arange(p1, p2 + dp, dp):
    x = xc + r * np.cos(p)
    y = yc + r * np.sin(p)
    if p > 90 * np.pi / 180 and p < 270 * np.pi / 180:
        plt.plot([xlast, x], [ylast, y], color='g', linestyle=':')
    else:
        plt.plot([xlast, x], [ylast, y], color='g')
    xlast = x
    ylast = y

plt.scatter(xc, yc, s=15, color='g')

# _____________________________________________red circle
xc = -20
yc = -20
r = 10

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100
xlast = xc + r * np.cos(p1)
ylast = yc + r * np.sin(p1)
for p in np.arange(p1, p2 + dp, dp):
    x = xc + r * np.cos(p)
    y = yc + r * np.sin(p)
    plt.plot([xlast, x], [ylast, y], linewidth=4, color='r')
    xlast = x
    ylast = y

plt.scatter(xc, yc, s=15, color='r')

# _____________________________________________purple circle
xc = 20
yc = 20
r = 50

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100
xlast = xc + r * np.cos(p1)
ylast = yc + r * np.sin(p1)
for p in np.arange(p1, p2 + dp, dp):
    x = xc + r * np.cos(p)
    y = yc + r * np.sin(p)
    plt.plot([xlast, x], [ylast, y], linewidth=2, color=(.8, 0., .8))
    xlast = x
    ylast = y

plt.scatter(xc, yc, color=(.5, 0., .5))

# _____________________________________________blue disc
xc = -53
yc = -30
r1 = 0
r2 = 10
dr = 1

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100
xlast = xc + r1 * np.cos(p1)
ylast = yc + r1 * np.sin(p1)
for r in np.arange(r1, r2, dr):
    for p in np.arange(p1, p2 + dp, dp):
        x = xc + r * np.cos(p)
        y = yc + r * np.sin(p)
        plt.plot([xlast,x],[ylast,y], linewidth=2, color=(0,0,.8))
        xlast = x
        ylast = y

plt.show()
```

```python
"""
ELLIPSES
"""

import numpy as np
import matplotlib.pyplot as plt

plt.axis([-75, 75, 50, -50])

plt.axis('on')
plt.grid(True)

plt.arrow(0, 0, 60, 0, head_length=4, head_width=3, color='k')
plt.arrow(0, 0, 0, 45, head_length=4, head_width=3, color='k')

plt.text(58, -3, 'x')
plt.text(-5, 44, 'y')

# _____________________________________________red ellipse
a = 40
b = 20.
p1 = 0
p2 = 180 * np.pi / 180
dp = .2 * np.pi / 180

xplast = a
yplast = 0
for p in np.arange(p1, p2, dp):
    xp = np.abs(a * b * (b * b + a * a * (np.tan(p))**2.)**-.5)
    yp = np.abs(a * b * (a * a + b / (np.tan(p)**2.))**-.5)
    if p > np.pi / 2:
        xp = -xp
    plt.plot([xplast, xp], [yplast, yp], color='r')
    plt.plot([xplast, xp], [-yplast, -yp], color='r')
    xplast = xp
    yplast = yp

# _____________________________________________green ellipse
a = 20.
b = 40.
xp1 = -a
xp2 = a
dx = .1

xplast = -a
yplast = 0
for xp in np.arange(xp1, xp2, dx):
    yp = b * (1 - xp**2. / a**2.)**.5
    plt.plot([xplast, xp], [yplast, yp], linewidth=1, color='g')
    plt.plot([xplast, xp], [-yplast, -yp], linewidth=1, color='g')
    xplast = xp
    yplast = yp

plt.plot([xplast,a],[yplast,0],linewidth=1,color='g')
plt.plot([xplast,a],[-yplast,0],linewidth=1,color='g')

# _____________________________________________blue ellipse
a = 5.
b = 15.
p1 = 0
p2 = 180 * np.pi / 180
dp = .2 * np.pi / 180

for p in np.arange(p1, p2, dp):
    xp = np.abs(a*b*(b*b+a*a*(np.tan(p))**2.)**-.5)
    yp = np.abs(a*b*(a*a+b*b/(np.tan(p)**2.))**-.5)
    if p > np.pi/2:
        xp = -xp
    plt.plot([xp, xp], [yp, -yp], linewidth=1, color='b')

plt.show()
```

```python
"""
2D TRANSLATION
"""

import numpy as np
import matplotlib.pyplot as plt

# Mendefinisikan batas sumbu
x1, x2 = -10, 140
y1, y2 = 90, -10

# Mengatur batas sumbu dan menyalakan grid
plt.axis([x1, x2, y1, y2])
plt.grid(True)
plt.axis('on')  # 'on' sebenarnya default dan bisa dihilangkan, tapi tetap ditampilkan sesuai maksud Anda

import matplotlib.pyplot as plt

# Menentukan batas area gambar
plt.axis([-10, 140, 90, -10])
plt.grid(True)
plt.title('Translation')

# Segitiga asli (hitam)
x = [20, 30, 40, 20]
y = [40, 20, 40, 40]
plt.plot(x, y, color='k', label='Asli')

# Translasi dx = 40 (hijau)
x_dx = [i + 40 for i in x]
plt.plot(x_dx, y, color='g', label='Translasi dx=40')

# Translasi dy = 40 (merah)
y_dy = [i + 40 for i in y]
plt.plot(x, y_dy, color='r', label='Translasi dy=40')

import numpy as np
import matplotlib.pyplot as plt

# Atur area dan grid
plt.axis([-10, 150, 60, 40])
plt.grid(True)
plt.title('Box Translation')

# Kotak awal (biru)
x_box = [0, 0, 5, 5, 0]
y_box = [55, 50, 50, 55, 55]
plt.plot(x_box, y_box, 'b')

# Translasi kotak secara horizontal (dx = 10 hingga dx = 120)
for dx in np.arange(10, 130, 10):  # Mulai dari 10 agar kotak awal tidak digambar dua kali
    x_translated = [x + dx for x in x_box]
    plt.plot(x_translated, y_box, 'b')

# Tampilkan hasil
plt.show()
```

```python
"""
2DROTRECTANGLE
"""

import matplotlib.pyplot as plt
import numpy as np

# Set batas axis dan aktifkan grid
plt.axis([-10, 150, 100, -10])
plt.grid(True)
plt.title('2D Rectangle Rotation')

# Gambar sumbu X dan Y
plt.arrow(0, 0, 40, 0, head_length=4, head_width=2, color='b')
plt.arrow(0, 0, 0, 40, head_length=4, head_width=2, color='b')
plt.text(30, -3, 'Xg', color='b')
plt.text(-8, 34, 'Yg', color='b')

# Pusat rotasi
xc, yc = 75, 50

# Gambar sumbu lokal rectangle (abu-abu)
plt.plot([xc - 40, xc + 60], [yc, yc], linewidth=1, color='grey')  # Sumbu X lokal
plt.plot([xc, xc], [yc - 40, yc + 45], linewidth=1, color='grey')  # Sumbu Y lokal
plt.text(127, 48, 'X')
plt.text(70, 90, 'Y')

# Tampilkan pusat rotasi
plt.scatter(xc, yc, s=20, color='k')  # titik pusat rotasi
plt.text(xc - 5, yc - 1, 'c')         # label titik pusat

# ------------------------------------------------------------
# Define function rotz (rotation about Z-axis in 2D)
def rotz(xp, yp, rz):
    c11 = np.cos(rz)
    c12 = -np.sin(rz)
    c21 = np.sin(rz)
    c22 = np.cos(rz)

    xpp = xp * c11 + yp * c12  # relative to xc, yc
    ypp = xp * c21 + yp * c22
    xg = xc + xpp              # transform to global Xg, Yg
    yg = yc + ypp
    return [xg, yg]

# ------------------------------------------------------------
# Define rectangle corner coordinates in local X,Y system
xp1, xp2, xp3, xp4 = -20, +20, +20, -20
yp1, yp2, yp3, yp4 = -5,  -5,  +5,  +5

# ------------------------------------------------------------
# Plot unrotated rectangle (local to global transformation)
plt.text(50, 45, '1')
plt.text(97, 45, '2')
plt.text(97, 57, '3')
plt.text(50, 57, '4')

# Plot rectangle corners before rotation
plt.scatter(xp1 + xc, yp1 + yc, s=10, color='k')
plt.scatter(xp2 + xc, yp2 + yc, s=10, color='k')
plt.scatter(xp3 + xc, yp3 + yc, s=10, color='k')
plt.scatter(xp4 + xc, yp4 + yc, s=10, color='k')

# Convert to global coordinates for connection plotting
xg1 = xc + xp1
yg1 = yc + yp1
xg2 = xc + xp2
yg2 = yc + yp2
xg3 = xc + xp3
yg3 = yc + yp3
xg4 = xc + xp4
yg4 = yc + yp4

# Draw the unrotated rectangle
plt.plot([xg1, xg2, xg3, xg4, xg1], [yg1, yg2, yg3, yg4, yg1], color='grey')

# Buat array koordinat rectangle sebelum rotasi
xg = [xg1, xg2, xg3, xg4, xg1]
yg = [yg1, yg2, yg3, yg4, yg1]

# Plot rectangle sebelum rotasi
plt.plot(xg, yg, color='k')

# ------------------------------------------------------------
# Rotasi rectangle (rotasi searah jarum jam)
rz = 45                      # derajat
rz = rz * np.pi / 180        # konversi ke radian

# ------------------- Point 1 -------------------
xp = xp1
yp = yp1
[xg, yg] = rotz(xp, yp, rz)
xg1, yg1 = xg, yg

# ------------------- Point 2 -------------------
xp = xp2
yp = yp2
[xg, yg] = rotz(xp, yp, rz)
xg2, yg2 = xg, yg

# ------------------- Point 3 -------------------
xp = xp3
yp = yp3
[xg, yg] = rotz(xp, yp, rz)
xg3, yg3 = xg, yg

# ------------------- Point 4 -------------------
xp = xp4
yp = yp4
[xg, yg] = rotz(xp, yp, rz)
xg4, yg4 = xg, yg

# ------------------------------------------------------------
# Plot rectangle setelah rotasi
xg_rot = [xg1, xg2, xg3, xg4, xg1]
yg_rot = [yg1, yg2, yg3, yg4, yg1]
plt.plot(xg_rot, yg_rot, color='r')  # warna merah untuk rectangle yang sudah dirotasi

# ------------------- Point 4 -------------------
xp = xp4
yp = yp4
[xg, yg] = rotz(xp, yp, rz)
xg4, yg4 = xg, yg

# ------------------- Plot rotated rectangle -------------------
xg_rot = [xg1, xg2, xg3, xg4, xg1]  # tutup kembali ke titik awal
yg_rot = [yg1, yg2, yg3, yg4, yg1]

plt.plot(xg_rot, yg_rot, color='r')  # rectangle setelah rotasi (warna merah)
plt.show()
```
